# SARSA: State-Action-Reward-State-Action

En este ejercicio vamos a implementar el algoritmo de SARSA como un ejemplo de los métodos on-policy para el aprendizaje por refuerzo. Esto es, crear agentes que aprenden a alcanzar un objetivo específico.

El método de SARSA se basa en el cálculo de los q-valores utilizando los valores calculados para el estado de llegada siguiendo la fórmula de actualización de los q-valores:

$Q(s,a) \leftarrow (1-\alpha)Q(s,a) + \alpha[R(s) + \gamma Q(s',a')] $

Para implementar SARSA definiremos un agente, `sarsa_agent.py` el cual utilizaremos para interactuar con el ambiente de Gridworld.

## Pregunta 1: ¿Cual es el comportamiento del agente con factor de descuento gamma = 1?

Comparamos el comportamiento del agente con gamma=0.96 (por defecto) vs gamma=1.0.

```python
# Entrenamiento con gamma = 1.0
np.random.seed(42)
env_g1 = CliffWalk()
agent_g1 = SARSA(env_g1, epsilon=0.9, gamma=1.0, alpha=0.81)
rewards_g1 = agent_g1.train(num_episodes=2000)

print("=== SARSA con gamma = 1.0 ===")
print("\nPolitica aprendida:")
agent_g1.print_policy()

print("Trayectoria greedy:")
path_g1 = agent_g1.print_path()
print(f"Longitud del camino: {len(path_g1) - 1} pasos")

# Comparar recompensas promedio
print("\n" + "=" * 60)
print("Comparacion de recompensas promedio (ultimos 100 episodios):")
print(f"  gamma = 0.96: {np.mean(rewards[-100:]):.1f}")
print(f"  gamma = 1.00: {np.mean(rewards_g1[-100:]):.1f}")
```

### Respuesta Pregunta 1

Con **gamma = 1.0** (sin descuento), el agente le da el mismo peso a recompensas futuras que a recompensas inmediatas. Esto produce los siguientes efectos:

1. **Camino mas conservador**: El agente tiende a alejarse aun mas del barranco. Con gamma=1, el costo total del camino largo (-13 pasos) no se descuenta, pero el costo de caer al barranco (-100) tampoco se descuenta. Dado que SARSA es on-policy y considera la probabilidad de exploracion, el agente internaliza completamente el riesgo de caer al barranco.

2. **Mayor sensibilidad al barranco**: Sin descuento, la penalizacion de -100 por caer al barranco se propaga con toda su magnitud a los estados cercanos. Con gamma=0.96, esta penalizacion se atenua ligeramente en cada paso, lo que puede hacer que el agente se acerque un poco mas al borde.

3. **Posible inestabilidad**: Con gamma=1 los Q-valores pueden crecer sin limite en episodios largos (la suma de -1 por paso no converge si el agente no llega al objetivo). Esto puede causar inestabilidad en el aprendizaje.

Con gamma=1, el agente busca minimizar el costo total del episodio sin descontar, lo que refuerza el comportamiento conservador de SARSA (camino seguro lejos del barranco).

## Pregunta 2: ¿Como podemos minimizar la trayectoria del agente?

Para minimizar la trayectoria, probamos:
1. **Aumentar epsilon** (mas explotacion, menos exploracion)
2. **Mas episodios de entrenamiento**
3. **Epsilon decreciente** (explorar al inicio, explotar despues)

```python
# Estrategia 1: Epsilon alto (0.99 - casi puramente greedy)
np.random.seed(42)
env1 = CliffWalk()
agent_high_eps = SARSA(env1, epsilon=0.99, gamma=0.96, alpha=0.81)
rewards_high_eps = agent_high_eps.train(num_episodes=2000)

print("=== Epsilon = 0.99 (casi greedy) ===")
agent_high_eps.print_policy()
print("Trayectoria:")
path1 = agent_high_eps.print_path()
print(f"Longitud del camino: {len(path1) - 1} pasos")

print("\n")

# Estrategia 2: Mas episodios de entrenamiento
np.random.seed(42)
env2 = CliffWalk()
agent_more = SARSA(env2, epsilon=0.9, gamma=0.96, alpha=0.81)
rewards_more = agent_more.train(num_episodes=5000)

print("=== 5000 episodios (vs 2000 por defecto) ===")
agent_more.print_policy()
print("Trayectoria:")
path2 = agent_more.print_path()
print(f"Longitud del camino: {len(path2) - 1} pasos")
```

```python
# Estrategia 3: Epsilon decreciente
np.random.seed(42)
env3 = CliffWalk()
agent_decay = SARSA(env3, epsilon=0.9, gamma=0.96, alpha=0.81)

# Entrenar con epsilon decreciente
total_episodes = 1000
rewards_decay = []
for ep in range(total_episodes):
    # Epsilon decae de 0.5 a 0.99 (mas exploracion al inicio, mas explotacion al final)
    agent_decay.epsilon = 0.5 + 0.49 * (ep / total_episodes)
    total_reward, steps = agent_decay.run_episode()
    rewards_decay.append(total_reward)

agent_decay.epsilon = 1.0  # greedy para evaluacion
print("=== Epsilon decreciente (0.5 -> 0.99) ===")
print(f"Epsilon final: {0.99:.2f}")
agent_decay.print_policy()
print("Trayectoria:")
path3 = agent_decay.print_path()
print(f"Longitud del camino: {len(path3) - 1} pasos")

# Resumen comparativo
print("\n" + "=" * 60)
print("Resumen: longitud de trayectoria greedy")
print(f"  Base (eps=0.9, 2000 eps):          {len(path) - 1} pasos")
print(f"  Epsilon alto (0.99):               {len(path1) - 1} pasos")
print(f"  Mas episodios (5000):              {len(path2) - 1} pasos")
print(f"  Epsilon decreciente (0.5->0.99):   {len(path3) - 1} pasos")
```

### Respuesta Pregunta 2

Para minimizar la trayectoria del agente entre el estado inicial y el estado de llegada, las estrategias mas efectivas son:

1. **Aumentar epsilon (mas explotacion)**: Con epsilon=0.99, el agente explota casi siempre la mejor accion conocida. Esto produce un camino mas corto porque SARSA, al ser on-policy, aprende Q-valores que reflejan el comportamiento real del agente. Si el agente explora poco, los Q-valores near the cliff son menos "contaminados" por el riesgo de caer, y el camino aprendido puede ser mas directo.

2. **Epsilon decreciente**: Esta es la mejor estrategia. Al inicio, epsilon bajo permite explorar ampliamente el espacio de estados (descubrir caminos). Gradualmente, al subir epsilon, el agente converge a explotar el mejor camino encontrado. Esto combina lo mejor de ambos mundos: buena exploracion y convergencia a un camino corto.

3. **Mas episodios**: Con mas entrenamiento, el agente refina mejor los Q-valores, pero dado que SARSA es on-policy, el camino sigue siendo conservador (lejos del barranco) porque los Q-valores reflejan el riesgo de exploracion.

**Conclusion clave**: La naturaleza on-policy de SARSA hace que siempre tienda a un camino seguro (no el mas corto). Para obtener el camino optimo (justo al borde del barranco), se necesitaria un metodo **off-policy** como Q-Learning, que aprende la politica optima independientemente de la politica de exploracion.
